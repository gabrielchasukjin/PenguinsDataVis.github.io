```python
dataset = { 
    
    "nodes":[],
    
    "edges": []

}
```

```python
def penguin_node(dictionary, x_value):
    
    node = {"Species" : penguins.iloc[x_value]['Species'], "Island" : penguins.iloc[x_value]['Island'], "Body Mass" : penguins.iloc[x_value]["Body Mass (g)"] / 2700}
    dictionary.append(node)
```

```python
for x in range(342): 
    penguin_node(dataset['nodes'],x)
```

```python
dataset['nodes'].append({"Species": "Torgersen Island", "Island": "Torgersen" ,"Body Mass": 3})
dataset['nodes'].append({"Species": "Dream Island", "Island": "Dream" ,"Body Mass": 3})
dataset['nodes'].append({"Species": "Biscoe Island", "Island": "Biscoe" ,"Body Mass": 3})
```

```python
def penguin_edges(dictionary, x_value):
    source_dict = {"Torgersen": 342, "Dream": 343, "Biscoe":344} 
    edge = {"source": x_value, "target": source_dict[penguins.iloc[x_value]['Island']]}
    dictionary.append(edge)
    
```

```python
for x in range(342): 
    penguin_edges(dataset['edges'],x)
```
